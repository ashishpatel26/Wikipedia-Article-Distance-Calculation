```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
### helping function
def cosine_similarity(x, y):
    return np.dot(x, y) / (np.sqrt(np.dot(x, x)) * np.sqrt(np.dot(y, y)))

def euclidean_distance(x, y):   
    return np.sqrt(np.sum((x - y) ** 2))

def l1_normalize(v):
    norm = np.sum(v)
    return v / norm

def l2_normalize(v):
    norm = np.sqrt(np.sum(np.square(v)))
    return v / norm
```

```python
corpus = ['Amitabh Bachchan','Salman khan','Akshay Kumar','Ranvir Singh','Ranvir Kapoor','Anil Ambani','Ratan tata','Amit Saghavi',
    'Kumar Manglam Birla','Azim premji','Lionel Messi','Roger federer','Sachin Tendulkar','Virat kohli','Mary kom','Bill Gates',
    'Steve Job','Mark Zuckerberg','Alan Turing','larry page']

def find_distance(n=5, data = corpus, Dist_function = euclidean_distance):
    """
    corpus = list which you want to pass
    Dist_function = Function Euclidean or Cosinesimilarity
    n = Number of Minimum distance you want to return
    """
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(data)
    X = np.array(X.todense())
    pair = []
    distance = []
    for i in range(len(corpus)):
        for j in range(len(corpus)-1):
            pair.append(str(corpus[i]+"-"+corpus[j]))
            distance.append(Dist_function(X[i],X[j]))
    df = pd.DataFrame()
    df['Pair'] = pair
    df['Distance'] = distance
    df.sort_values(by="Distance", inplace = True)
    print("Return Top {} Minimum Distance".format(n))
    return df[df['Distance'] > 0].head(n)
```

```python
find_distance(5, data=corpus, Dist_function = euclidean_distance)
```
